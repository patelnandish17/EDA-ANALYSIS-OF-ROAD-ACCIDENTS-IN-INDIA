```python
# EDA Script for Accident Prediction Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("accident_prediction_india.csv")

# Display first few rows
print("ðŸ”¹ First 5 Rows:")
print(df.head())

# Shape of the dataset
print("\nðŸ”¹ Dataset Shape:", df.shape)

# Check for null values
print("\nðŸ”¹ Missing Values:")
print(df.isnull().sum())

# Data types
print("\nðŸ”¹ Data Types:")
print(df.dtypes)

# Basic statistics
print("\nðŸ”¹ Statistical Summary:")
print(df.describe(include='all'))

# -------------------------------
# 1. Accident Severity Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Accident Severity', order=df['Accident Severity'].value_counts().index)
plt.title('Accident Severity Distribution')
plt.xlabel('Severity')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# 2. Top 10 States with Most Accidents
plt.figure(figsize=(10, 6))
top_states = df['State Name'].value_counts().head(10)
sns.barplot(x=top_states.values, y=top_states.index)
plt.title('Top 10 States with Most Accidents')
plt.xlabel('Number of Accidents')
plt.ylabel('State')
plt.tight_layout()
plt.show()

# 3. Accidents by Day of Week
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Day of Week', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Accidents by Day of the Week')
plt.tight_layout()
plt.show()

# 4. Accidents by Hour of Day
df['Hour'] = pd.to_datetime(df['Time of Day'], format='%H:%M', errors='coerce').dt.hour
plt.figure(figsize=(10, 4))
sns.histplot(df['Hour'].dropna(), bins=24, kde=True)
plt.title('Accidents by Hour of the Day')
plt.xlabel('Hour')
plt.tight_layout()
plt.show()

# 5. Vehicle Type Involved
plt.figure(figsize=(12, 5))
vehicle_types = df['Vehicle Type Involved'].value_counts()
sns.barplot(x=vehicle_types.index, y=vehicle_types.values)
plt.title('Vehicle Types Involved in Accidents')
plt.xlabel('Vehicle Type')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 6. Alcohol Involvement
plt.figure(figsize=(5, 4))
sns.countplot(data=df, x='Alcohol Involvement')
plt.title('Alcohol Involvement in Accidents')
plt.tight_layout()
plt.show()

# 7. Weather Conditions During Accidents
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x='Weather Conditions', order=df['Weather Conditions'].value_counts().index)
plt.title('Weather Conditions During Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 8. Lighting Conditions
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x='Lighting Conditions', order=df['Lighting Conditions'].value_counts().index)
plt.title('Lighting Conditions at Time of Accident')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 9. Road Condition
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Road Condition', order=df['Road Condition'].value_counts().index)
plt.title('Road Conditions During Accidents')
plt.tight_layout()
plt.show()

# 10. Correlation Heatmap (numeric data)
plt.figure(figsize=(8, 6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Numeric Features')
plt.tight_layout()
plt.show()
```
